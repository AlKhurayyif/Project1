#### Q1: How to find the hight of station with standerd error ..?

##### Firstly, I will find mean of AirTemperatureMean by Month:

```python
Temperature = df.groupby(['Year', 'Month']).agg({'AirTemperatureMeanDB':'mean'}).reset_index()
```

```python
Temperature
```

```python
Temperature = Temperature.pivot(index='Year', columns='Month', values='AirTemperatureMeanDB')
Temperature.columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Temperature = Temperature.reset_index()
```

```python
Temperature
```

###### Secondly, I will find mean of RelativeHumidityMean by Month:

```python
Humidity = df.groupby(['Year', 'Month']).agg({'RelativeHumidityMean': 'mean'}).reset_index()
```

```python
Humidity = Humidity.pivot(index='Year', columns='Month', values='RelativeHumidityMean')
Humidity.columns = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
Humidity = Humidity.reset_index()
```

```python
Humidity
```

###### Thirdly, I will calculate the annual average of temperature and humidity. Then, I will find the difference between each monthly average for both variables and the overall average:

```python
Temp_mean = df.groupby('Year').agg({'AirTemperatureMeanDB': 'mean'}).reset_index()
Temp_mean.columns = ['Year', 'AnnualTemperatureAverage']
Temp_mean
```

```python
Humi_mean = df.groupby('Year').agg({'RelativeHumidityMean': 'mean'}).reset_index()
Humi_mean.columns = ['Year', 'AnnualRelativeHumidityAverage']
Humi_mean
```

```python
Temp_delta = Temperature.copy()
Humi_delta = Humidity.copy()
```

```python
Temp_delta = Temp_delta.iloc[:, 1:] - Temp_mean['AnnualTemperatureAverage'].values[0]
Temp_delta['Year'] = Temp_mean['Year']
Temp_delta = Temp_delta[['Year', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']]
Temp_delta
```

```python
Humi_delta = Humi_delta.iloc[:, 1:] - Humi_mean['AnnualRelativeHumidityAverage'].values[0]
Humi_delta['Year'] = Humi_mean['Year']
Humi_delta = Humi_delta[['Year', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']]
Humi_delta
```

##### Finally, I will calculate the hight of Station above sea level by this Physical equation:

I created function called hight, It describe How to find the hight of station, I relied on these columns: PressureMeanStationLevel, PressureMeanSeaLevel and AirTemperatureMeanDB
To find it, and I will calculate the mean of hight.

```python
def Hight(df, PreStat_col, PreSea_col, Temp_col):
    
    '''
    ++++++++++++++++++++++++++++++++++
    |||     Barometric Formula     |||
    ++++++++++++++++++++++++++++++++++
    
        This Function besed on Barometric Formula, It describes the Relationshep between Hight and Pressure:
            dP/dh = -ρ_air * g 
        Where:
            ρ_air = (P * M_ / (R * T)    Density of air
            g = 9.80665 m / sec ** 2     Gravitational acceleration
            R =  8.314 J/K * mol         Universal gas constant
            M = 0.02896 K/mol            Molar mass of air
            T: Tempreture with Kelvin unit   [T_Kelin = 273.15 + T_Celsius]
            

        So, The General Equaton is:
                P(h_station) =  P(h=0) * exp(- M * g  h_station / (R * T))
            That means:
                h_station =  - ((R * T) / (M * g)) * ln(P(h=0) / P(h=h_station))
            Where:
                P(h=0) is Pressure at Sea Level

    
    
    ++++++++++++++++++++++++++++++++++++++
    |||     Parameters of function     |||
    ++++++++++++++++++++++++++++++++++++++
    
        df: DataFrame of my study
        PreStat_col: Column of Pressure at Station in df
        PreSea_col: Column of Pressure at Sea Level in df
        Temp_col: Column of Temperature at Station in df 

    
    
    ++++++++++++++++++++++
    |||     Return     |||
    ++++++++++++++++++++++
    
        h_Station in meters
    
    '''
    
    # Physical Constent:-
    g = 9.80665
    M = 0.02896
    R = 8.314
    
    # Claculation:-
    h_station = (R * (df[Temp_col] + 273.15) / (M * g)) * np.log(df[PreSea_col] / df[PreStat_col])
    
    return h_station
    
```

```python
df.columns
```

```python
df['Hight'] = Hight(df, 'PressureMeanStationLevel', 'PressureMeanSeaLevel', 'AirTemperatureMeanDB')
```

##### The Mean of Hight and Standard error

```python
# The Mean of Hight (meter):
Hight_mean = df['Hight'] .mean()
Hight_mean
```

```python
# The Standard error of Hight (meter):
Hight_sem = df['Hight'] .sem()
Hight_sem
```

#### Q3. Is There correlation between Air Temperature and Humidity by Month ..?

Now I wil make a new dataframe for find carrolation between Tempreture and Humidity.

```python
Temperature
```

```python
df_comp = Temperature.merge(Humidity, on='Year')  # That is for merge between Tempreture and Humidity.
df_comp.columns = ['Year', 'Jan_Temp', 'Feb_Temp', 'Mar_Temp', 'Apr_Temp', 'May_Temp', 'Jun_Temp', 'Jul_Temp',
       'Aug_Temp', 'Sep_Temp', 'Oct_Temp', 'Nov_Temp', 'Dec_Temp', 'Jan_Humi', 'Feb_Humi', 'Mar_Humi',
       'Apr_Humi', 'May_Humi', 'Jun_Humi', 'Jul_Humi', 'Aug_Humi', 'Sep_Humi', 'Oct_Humi', 'Nov_Humi',
       'Dec_Humi']
```

```python
df_comp = df_comp[[
    'Year', 'Jan_Temp',  'Jan_Humi', 'Feb_Temp', 'Feb_Humi', 'Mar_Temp', 'Mar_Humi', 'Apr_Temp', 'Apr_Humi',
    'May_Temp', 'May_Humi', 'Jun_Temp', 'Jun_Humi', 'Jul_Temp', 'Jul_Humi', 'Aug_Temp', 'Aug_Humi', 'Sep_Temp',
    'Sep_Humi', 'Oct_Temp', 'Oct_Humi', 'Nov_Temp', 'Nov_Humi', 'Dec_Temp',  'Dec_Humi'
]]
```

```python
df_comp
```

```python
corr_dict = {}
for i in range(12):
    corr_dict[df_comp.columns[2*i+1][0:3]] = df_comp[[df_comp.columns[0], 
                                                  df_comp.columns[2*i + 1],
                                                  df_comp.columns[2*i + 2]]].corr().values[1][2]
```

```python
df_corr = pd.DataFrame.from_dict(corr_dict, orient='index').reset_index()
df_corr.columns = ['Month', 'corr(Temp, Humi)']
```

```python
df_corr # Now I have DataFrame for Correlation. 
```

```python
fig = px.line(df_corr, x='Month', y='corr(Temp, Humi)')
fig.add_hrect(
        y0=0, y1=0.2, line_width=0, 
        fillcolor="green", opacity=0.1)
fig.add_hrect(
        y0=0.2, y1=0.5, line_width=0, 
        fillcolor="green", opacity=0.3)
fig.add_hrect(
        y0=0.5, y1=0.8, line_width=0, 
        fillcolor="green", opacity=0.5)
fig.add_hrect(
        y0=0.8, y1=1, line_width=0, 
        fillcolor="green", opacity=0.7)
fig.add_hrect(
        y0=0, y1=-0.2, line_width=0, 
        fillcolor="red", opacity=0.1)
fig.add_hrect(
        y0=-0.2, y1=-0.5, line_width=0, 
        fillcolor="red", opacity=0.3)
fig.add_hrect(
        y0=-0.5, y1=-0.8, line_width=0, 
        fillcolor="red", opacity=0.5)
fig.add_hrect(
        y0=-0.8, y1=-1, line_width=0, 
        fillcolor="red", opacity=0.7)
fig.update_layout(
    title={
        'text': "The Correlation between Temperature and Humidity for Month (2008 - 2017)",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
fig.show()
```
